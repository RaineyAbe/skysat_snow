```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
# Generate coordinates for a dummy iceberg
radius = 1e3          
n_points = 10000 
angles = np.linspace(0, 2 * np.pi, n_points)
x = radius * np.cos(angles)
y = radius * np.sin(angles)
# Add noise
noise_amplitude = 200
noise_frequency = 10
smooth_noise = noise_amplitude * np.sin(noise_frequency * angles)
x_noise = x + smooth_noise
y_noise = y + smooth_noise

plt.plot(x, y, '-')
plt.plot(x_noise, y_noise, '-')
plt.show()
```
